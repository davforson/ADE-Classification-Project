```python
#Removing duplicates from dataset
df = df[~(df.duplicated() == True)]
```

```python
df.shape
```

```python
df['label'].value_counts()
```

```python
#import spacy library, preprocessing data by removing stop words and punctuations, and carrying out lemmatization
import spacy

nlp = spacy.load('en_core_web_lg')

def preprocess(text):
    doc = nlp(text)

    filtered_token = []
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_token.append(token.lemma_)

    return " ".join(filtered_token)
```

```python
#creating a new column for preprocessed data
df['text_preprocessed'] = df['text'].apply(preprocess)
```

```python

df.head()
```

```python
#importing gensim, an NLP-trained model with word vectors for text representation 
import gensim.downloader as api

wv = api.load('fasttext-wiki-news-subwords-300')
```

```python
#creating new vector column for using the preprocessed text column
df['text_vector'] = df['text_preprocessed'].apply(lambda x: wv.get_mean_vector(x))
```

```python
df.head()
```

```python
#value for dominating target varible (false : 0)
max_value = df['label'].value_counts()[0]
max_value
```

```python
#Oversampling technique to address class imbalance
label_1 = df[df['label']==1].sample(max_value, replace = True)
label_0 = df[df['label']==0].sample(max_value)
```

```python
#concatenating the two dataframes after oversampling for the minor class
balanced_data = pd.concat([label_1, label_0], axis = 0)
```

```python
balanced_data.shape
```

```python
balanced_data.head()
```

```python
balanced_data['label'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
#splitting dataset for training and testing 
X_train, X_test, y_train, y_test = train_test_split(balanced_data['text_vector'],balanced_data['label'], test_size = 0.2, random_state = 42, stratify = balanced_data['label'])
```

```python
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler

#Converting array into 2D 
X_train_2d = np.stack(X_train)
X_test_2d = np.stack(X_test)

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train_2d)
X_test_scaled = scaler.transform(X_test_2d)
```

```python
model = MultinomialNB()

model.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test_scaled)

print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators = 50, criterion = 'entropy')
model.fit(X_train_2d, y_train)
```

```python
y_pred = model.predict(X_test_2d)

print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (8,5))
sns.heatmap(cm, annot = True, fmt = 'd') 
plt.xlabel('Prediction')
plt.ylabel('Truth')
```
